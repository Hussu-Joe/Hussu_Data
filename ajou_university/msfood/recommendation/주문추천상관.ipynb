```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
corr = pd.read_csv("./corrdf12new.csv",sep=";",encoding="cp949",engine="python",index_col="Unnamed: 0")
goods = np.array(corr.columns)
wgoods = pd.read_csv("./wgoodsdf12.csv",sep=";",encoding="cp949",engine="python",index_col="Unnamed: 0")
pvalue = pd.read_csv("./pvaluedf12.csv",sep=";",encoding="cp949",engine="python",index_col="Unnamed: 0")
leng = pd.read_csv("./lengdf12.csv",sep=";",encoding="cp949",engine="python",index_col="Unnamed: 0")
cos = pd.read_csv("./cosdf12.csv",sep=";",encoding="cp949",engine="python",index_col="Unnamed: 0")
df1 = pd.read_csv("./12월주문자료.CSV",encoding="cp949",engine="python",sep=";")
df1["주문횟수"]=np.repeat(1,len(df1))
df1["FC명"]=  [x.replace("<","").replace(">","") for x in df1.FC명]
df11 = pd.DataFrame({"SKU명":goods})
df11=df11.merge(df1.loc[:,["SKU명","SKU코드"]].drop_duplicates(),on="SKU명",how="left")
```

```python
only = df1[df1.위탁구분명!="범용"]
every = df1[df1.위탁구분명=="범용"]
```

```python
df1.SKU명.unique()
```

```python
def rec(x, fc):
    df2 = only[only["FC명"]==fc].append(every)
    df3 = df2.loc[:,["SKU명","SKU코드","소분류명","위탁구분명"]]
    df3 = df3.drop_duplicates()
    df4 = pd.DataFrame({"a":corr[x],"b":pvalue[x],"c":leng[x],"d":cos[x]})
    df4.columns = ["corr","pvalue","freq","cos"]
    df4["SKU명"]= wgoods[x]
    df5 = df3.merge(df4,on="SKU명",how="left")
    df6 = df5.dropna(subset=["corr"])
    df7 = df6[df6["corr"]>=0].sort_values("corr",ascending=False)
    df8 = df7[df7["pvalue"]<=0.10]
    df9 = df8[df8["freq"]>=5]
    cat1 = df3.소분류명[df3.SKU명==x].unique()[0]
    cats = df9[df9.소분류명==cat1]
    catd = df9[df9.소분류명!=cat1]
    return(cats, catd)
```

```python
s, d=rec('매일우유 오리지널(흰우유)1000ML*12', '매일유업_폴바셋')
```
