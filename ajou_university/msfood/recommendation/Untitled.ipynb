```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import seaborn as sns
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, euclidean
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
```

```python
df1 = pd.read_csv("./12월주문자료.CSV",encoding="cp949",engine="python",sep=";")
df1["주문횟수"]=np.repeat(1,len(df1))
ttt=  [x.replace("<","").replace(">","") for x in df1.FC명]
df1["FC명"]=ttt
```

```python
order = df1.pivot_table(index="매출처명",columns="SKU명",values="주문횟수",aggfunc=np.sum)
```

```python
quant = df1.pivot_table(index="매출처명",columns="SKU명",values="주문수량(수불단위)",aggfunc=np.sum)
```

```python
aa=pd.DataFrame(order.sum())
```

```python
aa.columns =["주문횟수"]
```

```python
aa["주문수량"]=quant.sum()
```

```python
plt.plot(np.log(aa["주문수량"]),np.log(aa["주문횟수"]),".")
plt.ylabel("number of order")
plt.xlabel("quantity")
plt.xticks(fontsize=8,rotation=45)
plt.title("orders quant")
plt.savefig("./주문수량횟수.png",dpi=400)
```
