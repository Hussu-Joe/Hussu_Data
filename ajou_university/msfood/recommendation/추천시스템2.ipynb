```python
matrix=df2.pivot_table(index="매출처명",columns="상품명",values="단수",aggfunc=np.sum)
```

```python
흰설탕=matrix['흰설탕(삼양사)15kg']
similar=matrix.corrwith(흰설탕)
corr흑 = pd.DataFrame(similar, columns=["Correlation"])
corr흑.dropna(inplace=True)
corr흑=corr흑.sort_values("Correlation",ascending=False)
corr흑 = corr흑.join(count)
```
